```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the results of the simulation for the homogeneous settin
p_r = 50 # or 100
L_r = 10
gamma_seq = np.linspace(0.45, 0.95, 20)
res_all = np.zeros((len(gamma_seq), 2))
for i in range(len(gamma_seq)):
    gamma_r = gamma_seq[i]
    res = np.loadtxt(f"results/res_p={p_r}_L={L_r}_gamma={gamma_r}.txt")
    res_all[i, 0] = np.mean(res[:, 0])
    res_all[i, 1] = np.mean(res[:, 1])

res_all_10 = res_all

L_r = 20
gamma_seq = np.linspace(0.45, 0.95, 20)
res_all = np.zeros((len(gamma_seq), 2))
for i in range(len(gamma_seq)):
    gamma_r = gamma_seq[i]
    res = np.loadtxt(f"results/res_p={p_r}_L={L_r}_gamma={gamma_r}.txt")
    res_all[i, 0] = np.mean(res[:, 0])
    res_all[i, 1] = np.mean(res[:, 1])

res_all_20 = res_all
```

```python
# Plot the results of the simulation for the homogeneous setting (Figure 1)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'font.size': 18})
ms = 5
lw = 1.15
x_seq = np.linspace(0.45, 0.95, 20)
plt.plot(x_seq, res_all_10[:, 0], linewidth=lw, marker='o', linestyle='-', markersize=ms, label="distributed $L=10$", color="blue")
plt.plot(x_seq, res_all_10[:, 1], linewidth=lw, marker='s', linestyle='-', markersize=ms, label="pooled $L=10$", color="red")
plt.plot(x_seq, res_all_20[:, 0], linewidth=lw, marker='o', linestyle='--', markersize=ms, label="distributed $L=20$", color="blue")
plt.plot(x_seq, res_all_20[:, 1], linewidth=lw, marker='s', linestyle='--', markersize=ms, label="pooled $L=20$", color="red")


plt.legend(loc='upper right', fontsize=15)
plt.xlabel(r"$\log_{p}(\lambda / \sigma)$")
plt.ylabel("Estimation Error")
plt.grid(True)

#plt.savefig(f"homo_p={p_r}_error.png", bbox_inches='tight', dpi=150)

plt.show()
```

```python
# Read the results of the simulation for the heterogeneous setting with the same core

p_r = 50 # or 100
L_r = 10
gamma_seq = np.linspace(0.45, 0.95, 20)
res_all = np.zeros((len(gamma_seq), 5))
for i in range(len(gamma_seq)):
    gamma_r = gamma_seq[i]
    res = np.loadtxt(f"results/hetero_res_p={p_r}_L={L_r}_gamma={gamma_r}.txt")
    res_all[i, :] = np.mean(res, axis=0)

res_all_10 = res_all

L_r = 20
gamma_seq = np.linspace(0.45, 0.95, 20)
res_all = np.zeros((len(gamma_seq), 5))
for i in range(len(gamma_seq)):
    gamma_r = gamma_seq[i]
    res = np.loadtxt(f"results/hetero_res_p={p_r}_L={L_r}_gamma={gamma_r}.txt")
    res_all[i, :] = np.mean(res, axis=0)

res_all_20 = res_all
```

```python
# Plot the results of the simulation for the heterogeneous setting with the same core (Figure 2)
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'font.size': 18})
ms = 5
lw = 1.15
x_seq = np.linspace(0.45, 0.95, 20)
plt.plot(x_seq, res_all_10[:, 2], linewidth=lw, marker='^', linestyle='-', markersize=ms, label="single", color="green")
plt.plot(x_seq, res_all_10[:, 0], linewidth=lw, marker='o', linestyle='-', markersize=ms, label="distributed $L=10$", color="blue")
plt.plot(x_seq, res_all_10[:, 4], linewidth=lw, marker='s', linestyle='-', markersize=ms, label="pooled $L=10$", color="red")
plt.plot(x_seq, res_all_20[:, 0], linewidth=lw, marker='o', linestyle='--', markersize=ms, label="distributed $L=20$", color="blue")
plt.plot(x_seq, res_all_20[:, 4], linewidth=lw, marker='s', linestyle='--', markersize=ms, label="pooled $L=20$", color="red")

plt.legend(loc='upper right', fontsize=12)
plt.xlabel(r"$\log_{p}(\lambda / \sigma)$")
plt.ylabel("Estimation Error")
plt.grid(True)
# plt.yscale("log")
# plt.savefig(f"hetero_p={p_r}_error.png", bbox_inches='tight', dpi=150)
plt.show()
```
